```python
import pandas as pd
```

```python
bikes_full = pd.read_csv('JC-201808-citibike-tripdata.csv')
bikes_full.head()
```

```python
bikes_start = bikes_full[['start station name','start station latitude','start station longitude']]
bikes_start.head()
```

```python
bikes_start['position']='start'
bikes_start.head()
```

```python
bikes_start=bikes_start.rename(columns={'start station name':'station name', 'start station latitude':"station latitude",'start station longitude':'station longitude'})
```

```python
bikes_start.count()
```

```python
bikes_end = bikes_full[['end station name','end station latitude','end station longitude']]
bikes_end.head()
```

```python
bikes_end['position']='end'
bikes_end.head()
```

```python
bikes_end=bikes_end.rename(columns={'end station name':'station name', 'end station latitude':"station latitude",'end station longitude':'station longitude'})
bikes_end.count()
```

```python
js_bikes = pd.concat([bikes_start, bikes_end], ignore_index=True)
js_bikes.count()
```

```python
js_bikes.head()
```

```python
ny_bikes_full = pd.read_csv('201808-citibike-tripdata.csv')
ny_bikes_full.dropna(how='any')
```

```python
ny_bikes_start = ny_bikes_full[['start station name','start station latitude','start station longitude']]
ny_bikes_start['position']='start'
ny_bikes_end = ny_bikes_full[['end station name','end station latitude','end station longitude']]
ny_bikes_end['position']='end'

ny_bikes_end.head()
```

```python
ny_bikes_start=ny_bikes_start.rename(columns={'start station name':'station name', 'start station latitude':"station latitude",'start station longitude':'station longitude'})
```

```python
ny_bikes_end=ny_bikes_end.rename(columns={'end station name':'station name', 'end station latitude':"station latitude",'end station longitude':'station longitude'})
```

```python
ny_bikes = pd.concat([ny_bikes_start, ny_bikes_end], ignore_index=True)
ny_bikes.head()
```

```python
js_ny_bikes = pd.concat([ny_bikes, js_bikes], ignore_index=True)
```

```python
js_ny_bikes.to_csv('js_ny_bikes.csv', index=False)
```
